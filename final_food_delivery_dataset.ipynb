```python
import pandas as pd
import sqlite3
import json
```

```python
orders_df=pd.read_csv("C://Users//HP//Downloads//orders.csv")
```

```python
with open("C://Users//HP//Downloads//users.json",'r') as f:
    users_data=json.load(f)
users = pd.DataFrame(users_data)
```

```python
conn=sqlite3.connect(':memory:')
with open("C://Users//HP//Downloads//restaurants.sql",'r') as f:
    conn.executescript(f.read())
restaurants=pd.read_sql_query("SELECT * FROM restaurants",conn)
```

```python
merged_data=pd.merge(orders_df,users,on='user_id',how='left')
```

```python
final_df=pd.merge(merged_data, restaurants,on='restaurant_id',how='left',suffixes=('_order','_master'))
```
